# Data Visualization for COVID-19 Cases and Deaths in Europe

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df[df['continent'] == 'Europe'], x='location',y='total_cases', label='Total Cases')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Number of cases')
plt.title('Data Visualization for COVID-19 Cases in Europe')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df[df['continent'] == 'Europe'], x='location', y='total_deaths', label='Total Deaths',color='orange')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Number of deaths')
plt.title('Data Visualization for COVID-19  Deaths in Europe ')
plt.legend()
plt.show()

```

```python
correlation_value=df['total_cases'].corr(df['total_deaths'])
print(f"Correlation Between Total Cases and Total Deaths : {correlation_value}")
```

```python
df_euro = df[df['continent'] == 'Europe'][['location', 'total_cases', 'total_vaccinations']].dropna()
correlation = df_euro['total_cases'].corr(df_euro['total_vaccinations'])
print(f"Correlation: {correlation:.2f}")


plt.figure(figsize=(12, 6))

sns.lineplot(data=df_euro, x='location', y='total_cases', label='Total Cases', marker='o', linestyle='-')
sns.lineplot(data=df_euro, x='location', y='total_vaccinations', label='Total Vaccinations', marker='s', linestyle='--')

plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Total Number')
plt.title('Relationship Between Vaccination and Case Numbers in Europe')
plt.legend()
plt.grid(True)
plt.show()


```

# Correlation and Visualization of the Relationship Between Population Aged 65 and Older and Total Deaths"

```python
df_cleaned = df[['aged_65_older', 'total_deaths']].dropna()
correlation = df_cleaned['aged_65_older'].corr(df_cleaned['total_deaths'])
print(f"Correlation Between Population Aged 65 and Older and Total Deaths: {correlation}")
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df_cleaned.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("65 Yaş Üstü ve Ölüm Oranı Korelasyon Matrisi")
plt.show()
```

# Relationship Between Vaccination Numbers and Case Numbers on Specific Days in Turkey (2021)

```python
df_Türkiye=df[df['location']=='Turkey']
```

```python
df_Türkiye = df_Türkiye[
    ['continent', 'location', 'date', 'total_cases', 'total_deaths','total_vaccinations']
]
```

```python
df_Türkiye = df_Türkiye.drop_duplicates(subset=['total_cases'])
df_Türkiye.shape
```

```python
df_Türkiye.head()
```

```python


df_Türkiye['date'] = pd.to_datetime(df_Türkiye['date'])
df_Türkiye = df_Türkiye[df_Türkiye['date'].dt.year.isin([2020, 2021])]
df_Türkiye['daily_cases'] = df_Türkiye['total_cases'].diff().fillna(0)
df_Türkiye['daily_vaccinations'] = df_Türkiye['total_vaccinations'].diff().fillna(0)

fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.bar(df_Türkiye['date'].astype(str), df_Türkiye['daily_cases'], width=0.4, label='Daily Cases', color='silver')

ax2 = ax1.twinx()
ax2.plot(df_Türkiye['date'].astype(str), df_Türkiye['daily_vaccinations'], label='Daily Vaccinations', color='green', marker='o', linewidth=2)

ax1.set_xlabel('Date')
ax1.set_ylabel('Daily Cases', color='silver')
ax2.set_ylabel('Daily Vaccinations', color='green')

plt.title('COVID-19 Cases and Vaccinations in Turkey (2021)')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

ax1.set_xticklabels(df_Türkiye['date'].dt.strftime('%Y-%m-%d'), rotation=90)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
